# -7-Filtrage et indexation avancée et Tri


```python
B = np.array([[5, 15, 25],
[35, 45, 55],
[65, 75, 85]])

lines = np.array([True, False, True])
print(B[lines, :])

col = np.array([False, True, False])
print(B[:, col])

elements = np.array([[True, False, False],
[False, False, False],
[False, False, True]])
print(B[elements])
```

```python
A = np.array([[5, 10, 25],
[35, 45, 55],
[65, 75, 80]])
print(np.where(A % 2 == 0))
print(np.where(A % 2 == 0, A, 0))
```
