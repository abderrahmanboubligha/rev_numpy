```python
L=['a', 'b', 'c','a','c','c']
d = {i: L.count(i) for i in set(L)}
print(d)

d={}
for i in L:
    if i in d:
        d[i] += 1
    else:
        d[i] = 1
print(d)


#white get method
d={}
for elt in L:
    d[elt] = d.get(elt, 0) + 1
print(d)

```

```python
list_mots = ["hello", "hiiiiii", "bonjour", "salama alaikom", "java", "python"]
dictionnaire2={}
for mot in list_mots:
    if len(mot) in dictionnaire2:
        dictionnaire2[len(mot)].append(mot)
    else:
        dictionnaire2[len(mot)] = [mot]
print(dictionnaire2)

dictionnaire3={}
for mot in list_mots:
    dictionnaire3[len(mot)]=dictionnaire3.get(len(mot), [])
    dictionnaire3[len(mot)].append(mot)
    
print(dictionnaire3)
```
